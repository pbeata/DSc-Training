### Machine Learning Application

We will use a logistic regression in this analysis. 

```python
import numpy as np
import pandas as pd
```

```python
data_preprocessed = pd.read_csv('./data/absenteeism_preprocessed_download.csv')
data_preprocessed.head()
```

### Create the Targets

```python
# we will use the median of absent time to determine the cut-off for "excessive" time missed
median_absent_time = data_preprocessed['Absenteeism Time in Hours'].median()
median_absent_time
```

```python
# np.where() is like the Excel IF function: (condition, if true do this, else false do this)
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > median_absent_time, 1, 0)
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
data_preprocessed.head()
```

Using the median to define our targets into two groups, we implicitly balanced the dataset already! We expect roughly half of the targets to 0 and the other half to be 1.

```python
# in Logistic regression, a 45-55 split (or even 40-60) is acceptable for balance, but with NNs it is not
print(targets.sum() / targets.shape[0])
print(1 - targets.sum() / targets.shape[0])
```

```python
# CHECKPOINT 1

# data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours'], axis=1)
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours', 
                                            'Day of the Week',
                                            'Daily Work Load Average',
                                            'Distance to Work'], axis=1)

data_with_targets is data_preprocessed
```

```python
data_with_targets.head()
```

### Select the Inputs for Regression

```python
data_with_targets.shape
```

```python
# use all columns EXCEPT the last one (targets)
# data_with_targets.iloc[:, 0:14]
data_with_targets.iloc[:, :-1]
```

```python
unscaled_inputs = data_with_targets.iloc[:, :-1]
unscaled_inputs.columns.values
```

```python
# standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
# CREATE A CUSTOM SCALER TO AVOID STANDARDIZING THE DUMMY VARIABLES

# (note: in practice, we would standardize before creating the dummy variables)

from sklearn.base import BaseEstimator, TransformerMixin

class CustomScaler (BaseEstimator, TransformerMixin):
    
#     def __init__ (self, columns, copy=True, with_mean=True, with_std=True):
    def __init__ (self, columns):
#         self.scaler = StandardScaler(copy, with_mean, with_std)
        self.scaler = StandardScaler()
        self.columns = columns
        self.mean_ = None
        self.var_ = None
    
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self

    def transform(self, X, y=None, copy=None):
        init_column_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_unscaled = X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_unscaled, X_scaled], axis=1)[init_column_order]
```

```python
# remove the dummy variables so that the custom scaler does not scale those features

# columns_to_scale = ['Month Value',
#        'Day of the Week', 'Transportation Expense', 'Distance to Work',
#        'Age', 'Daily Work Load Average', 'Body Mass Index',
#        'Children', 'Pet']

columns_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']
columns_to_scale = [x for x in unscaled_inputs.columns.values if x not in columns_to_omit]
```

```python
# scaler = StandardScaler()
scaler = CustomScaler(columns_to_scale)
```

```python
scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = scaler.transform(unscaled_inputs)
scaled_inputs
```

```python
scaled_inputs.shape
```

### Splitting the Data for Training and Testing

```python
# splitting and shuffling
```

```python
from sklearn.model_selection import train_test_split
```

```python
# sklearn is randomly shuffling automatically, but we can set the random_state to make sure it is the same shuffle each time
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size=0.8, random_state=20)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

### Logistic Regression with sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
# train the model
reg = LogisticRegression()
reg.fit(x_train, y_train)
```

```python
# check model accuracy
reg.score(x_train, y_train)
```

```python
# manually check the accuracy
model_outputs = reg.predict(x_train)

model_outputs # these are the predictions of our model
```

```python
model_outputs == y_train
```

```python
np.sum(model_outputs == y_train)  # how many are true
```

```python
# accuracy
np.sum(model_outputs == y_train) / len(model_outputs)
```

### Getting the Incercepts and Coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
# get all feature names
feature_names = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature'], data=feature_names)

summary_table['Coefficient'] = np.transpose(reg.coef_)

summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table
```

```python
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds Ratio'] = np.exp(summary_table.Coefficient)
summary_table
```

```python
summary_table.sort_values('Odds Ratio', ascending=False)

# RE-WATCH THIS LECTURE ON ODDS RATIO (VIDEO 451)
```

### Testing our ML Model

```python
round(100 * reg.score(x_test, y_test), 1)
```

```python
predicted_probs = reg.predict_proba(x_test)
predicted_probs[:, 1]
```
