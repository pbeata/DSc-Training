### TASK: Recreate the Scatter Plot shown below

**The scatterplot attempts to show the relationship between the days employed versus the age of the person (DAYS_BIRTH) for people who were not unemployed. Note, to reproduce this chart you must remove unemployed people from the dataset first. Also note the sign of the axis, they are both transformed to be positive.
```python
df_employed = df[ df['OCCUPATION_TYPE'].notnull() ]
df_employed = df_employed.reset_index(drop=True)
df_employed['OCCUPATION_TYPE'].unique()
```

### TASK: Recreate the Categorical Plot

**This plot shows information only for the *bottom half* of income earners in the data set. It shows the boxplots for each category of NAME_FAMILY_STATUS column for displaying their distribution of their total income. Note: You will need to adjust or only take part of the dataframe *before* recreating this plot. You may want to explore the *order* parameter to get the xticks in the exact order shown here**

```python
q = df['AMT_INCOME_TOTAL'].quantile(0.5)
df2 = df[ df['AMT_INCOME_TOTAL'] < q ]
df2.describe()
```

```python
plt.figure(figsize=(10,3), dpi=300)
sns.boxplot(data=df2, y='AMT_INCOME_TOTAL', x='NAME_FAMILY_STATUS',
           hue='FLAG_OWN_REALTY');
plt.legend(loc=(1.05,0.5));
```

### Solution

```python
int(len(df)/2)
```

```python
df_bottom = df.nsmallest(n=int(len(df)/2), columns='AMT_INCOME_TOTAL')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df_bottom, y='AMT_INCOME_TOTAL', x='NAME_FAMILY_STATUS',
           hue='FLAG_OWN_REALTY');

plt.legend(loc=(1.05,0.5), title='FLAG_OWN_REALTY');
```
