```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%config Completer.use_jedi = False
```

```python
df = pd.read_csv('./raw_data/Intercity_Air__Bus__and_Rail_Transportation_Facilities.csv')
df.head()
```

```python
df.info()
```

```python
df['year'].unique()
```

```python
df['year'] = df['year'].astype(int)
```

```python
df_air = df[df['mode'] == 'Air']
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.scatterplot(x='Longitude', y='Latitude', data=df, 
                alpha=0.4, hue='mode')
plt.show()
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.scatterplot(x='Longitude', y='Latitude', data=df_air, 
                alpha=0.4, hue='year')
plt.show()
```

```python
sns.countplot(data=df, x='mode', hue='year');
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.countplot(data=df_air[df_air['year'] == 2018], x='State')
plt.xticks(rotation=90)
plt.show()
```
