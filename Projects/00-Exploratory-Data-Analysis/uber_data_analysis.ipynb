```python
def convert_date(date):
    # expected format: mm/dd
    # we automatically add the year (2021)
    tmp = date.split('/')
    year = '2021'
    return f"{tmp[1]-tmp[0]-year}"

# my_date = '3/21'
# convert_date(my_date)
```

```python
def add_year(orig_date):
    return orig_date + '/2021'
```

```python
df['Date'] = df['Date'].apply(add_year)
df
```

```python
df['Datetime'] = pd.to_datetime(df['Date'])
df
```

```python
df['Day'] = df['Datetime'].dt.day_name()
```

```python
cat_cols = ['Day']
df_dum = pd.get_dummies(df[cat_cols], drop_first=True)
df_dum
```

```python
drop_cols = ['Week', 'Date', 'Datetime', 'Day_of_Week', 'Day', 'Hours', 'Minutes']
df_final = df.drop(drop_cols, axis=1)
df_final
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_final)
```

```python
from sklearn.cluster import KMeans
```

```python
K = 2
model = KMeans(n_clusters=K)
labels = model.fit_predict(X_scaled)
labels
```
