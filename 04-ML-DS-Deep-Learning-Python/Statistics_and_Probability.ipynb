```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# The mean is $27,000 and the standard deviation is $15,000 
# We take 10000 samples from the Normal distribution
incomes = np.random.normal(27000, 15000, 10000)
np.mean(incomes)
```

```python
plt.hist(incomes, bins=50)
plt.show()
```

```python
np.median(incomes)
```

```python
incomes = np.append(incomes, [1000000000])
```

```python
np.median(incomes)
```

```python
# notice how the outlier changes the mean by a lot!
np.mean(incomes)
```

## Data Distributions

```python
values = np.random.uniform(-10, 10, 100000)
plt.hist(values, 50)
plt.show()
```

```python
x = np.arange(-3, 3, 0.001)
plt.plot(x, stats.norm.pdf(x))
plt.show()
```

```python
mu = 5.0
sigma = 2.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

```python
x = np.arange(0, 10, 0.001)
plt.plot(x, stats.expon.pdf(x))
plt.show()
```

```python
n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, stats.binom.pmf(x, n, p))
plt.show()
```

```python
mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, stats.poisson.pmf(x, mu))
plt.show()
```

## Moments

* 1st Moment: Mean
* 2nd: Variance
* 3rd: Skew, how lopsided the distribution is (look at the tails)
* 4th: Kurtosis, how thick is the tail and how sharp is the peak compared to the normal distribution

```python
vals = np.random.normal(0, 0.5, 100000)
plt.hist(vals, 50)
# plt.xlim(0.5, 1.5)
plt.show()
```

```python
np.mean(vals)
```

```python
np.var(vals)
```

```python
np.std(vals)
```

```python
stats.skew(vals)
```

```python
stats.kurtosis(vals)
```
