```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('../data/admittance.csv')
raw_data
```

```python
# convert admitted column using dummies
data = raw_data.copy()
data['Admitted'] = data['Admitted'].map({'Yes':1, 'No':0})
data
```

```python
# variables
y = data['Admitted']
x1 = data['SAT']
```

```python
# scatter plot
plt.scatter(x1, y, color='C0')
plt.xlabel('SAT')
plt.ylabel('Admitted')
plt.show()
```

```python
# linear regression
x = sm.add_constant(x1)
reg_lin = sm.OLS(y, x)
results_lin = reg_lin.fit()
```

```python
plt.scatter(x1, y, color='C0')
yhat = x1 * results_lin.params[1] + results_lin.params[0]

plt.plot(x1, yhat, lw=2.5)
plt.xlabel('SAT')
plt.ylabel('Admitted')
plt.show()
```

Our data is clearly nonlinear.

```python
# now include SAT score as well
y = data['Admitted']
x1 = data[['SAT', 'Gender']]
```

```python
# REGRESSION
x = sm.add_constant(x1)
reg = sm.Logit(y, x)
results = reg.fit()
results.summary()
```

```python
results.predict()
```

```python
np.array(data['Admitted'])
```

```python
results.pred_table()
```

```python
cm = pd.DataFrame(results.pred_table())
cm.columns = ['Predicted 0', 'Predicted 1']
cm = cm.rename(index={0: 'Actual 0', 1: 'Actual 1'})
cm  # confusion matrix
```

This shows how confused our model was.

For 69 observations, the model predicted 0 when the true value was 0.

For 90 observations, the model predicted 1 when the true value was 1.

**SUMMARY:**
* the model was correct in 159 cases
* the model was incorrect in 5 + 4 = 9 cases (confused)

```python
# compute the accuracy
C = np.array(cm)
train_accuracy = (C[0,0] + C[1,1]) / C.sum()

print("The training accuracy of the model is {x}%.".format(x=round(100 * train_accuracy, 1)))
```

```python
# Expand the model to include more features
x1 = df[['interest_rate', 'march', 'credit', 'previous', 'duration']]
y = df['y'].map({'yes':1, 'no':0})

# logistic regression
x = sm.add_constant(x1)
reg = sm.Logit(y, x)
results = reg.fit()
results.summary2()
```

```python
# confusion matrix and accuracy

def confusion_matrix(data, actual_values, model):
         
        # Confusion matrix 
        
        # Parameters
        # ----------
        # data: data frame or array
            # data is a data frame formatted in the same way as your input data (without the actual values)
            # e.g. const, var1, var2, etc. Order is very important!
        # actual_values: data frame or array
            # These are the actual values from the test_data
            # In the case of a logistic regression, it should be a single column with 0s and 1s
            
        # model: a LogitResults object
            # this is the variable where you have the fitted model 
            # e.g. results_log in this course
        # ----------
        
        #Predict the values using the Logit model
        pred_values = model.predict(data)
        # Specify the bins 
        bins=np.array([0,0.5,1])
        # Create a histogram, where if values are between 0 and 0.5 tell will be considered 0
        # if they are between 0.5 and 1, they will be considered 1
        cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
        # Calculate the accuracy
        accuracy = (cm[0,0]+cm[1,1])/cm.sum()
        # Return the confusion matrix and 
        return cm, accuracy
```

```python
confusion_matrix(x, y, results)
```

```python
data = pd.read_csv('../data/binary_predictors.csv')
data['Admitted'] = data['Admitted'].map({'Yes':1, 'No':0})
data['Gender'] = data['Gender'].map({'Female':1, 'Male':0})
data
```

```python
# variables 
y = data['Admitted']
x1 = data[['SAT', 'Gender']]

# regression
x = sm.add_constant(x1)
reg = sm.Logit(y, x)
results = reg.fit()
```

```python
x
```

```python
test_actual = test['Admitted']  # observed outcomes
test_data = test.drop(['Admitted'], axis=1)
test_data = sm.add_constant(test_data)
test_data

# make sure to pay attention to COLUMN ORDER
```

```python
def confusion_matrix(data, actual_values, model):
    
    pred_values = model.predict(data)
    bins = np.array([0, 0.5, 1])
    cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
    accuracy = (cm[0,0] + cm[1,1]) / cm.sum()
    return cm, accuracy
```

```python
cm = confusion_matrix(test_data, test_actual, results)
cm
```

* The test accuracy shown here (89.4%) is the figure we use when we refer to overall accuracy of the regression. 
* The training accuracy is almost always higher than the test accuracy
