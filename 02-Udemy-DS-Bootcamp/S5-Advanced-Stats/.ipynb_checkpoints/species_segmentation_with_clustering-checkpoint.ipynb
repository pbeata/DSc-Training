Tasks for this investigation:
* Start by creating 2 clusters. 
* Then standardize the data and try again. Does it make a difference?
* Use the Elbow rule to determine how many clusters there really are.

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
raw_data = pd.read_csv('../data/iris.csv')
raw_data
```

```python
raw_data.describe()
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
plt.scatter(raw_data['sepal_length'], raw_data['sepal_width'])

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# plt.xlim(0, 10)
# plt.ylim(0, 6)

plt.show()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
kmeans = KMeans(2)
kmeans.fit(raw_data)
```

```python
cluster_data = raw_data.copy()
cluster_data['cluster'] = kmeans.fit_predict(raw_data)
cluster_data
```

```python
plt.scatter(raw_data['sepal_length'], raw_data['sepal_width'], c=cluster_data['cluster'], cmap='rainbow')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.show()
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
# standardize the variables
from sklearn import preprocessing
data_scaled = preprocessing.scale(raw_data)
data_scaled
```

# Clustering (scaled data)

```python
kmeans_scaled = KMeans(2)
kmeans_scaled.fit(data_scaled)
```

```python
data_scaled_clustered = pd.DataFrame(data_scaled)
data_scaled_clustered['cluster'] = kmeans_scaled.fit_predict(data_scaled)
data_scaled_clustered
```

```python
data_scaled_clustered.describe()
```

```python
plt.scatter(data_scaled_clustered[0], data_scaled_clustered[1], c=data_scaled_clustered['cluster'], cmap='rainbow')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.show()
```

## Take Advantage of the Elbow Method

### WCSS

```python
# calculate the WCSS for any number of clusters up to N
# WCSS: within clusters sum of squares

N = 10
WCSS = []
x_scaled = pd.DataFrame(data_scaled)

for i in range (1, N+1):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    WCSS.append(kmeans.inertia_)

WCSS
```

### The Elbow Method

```python
plt.figure(figsize=(6,3), dpi=100)

n = np.arange(1, N+1)
plt.plot(n, WCSS, 'ko-')
plt.ylabel('WCSS')
plt.xlabel('Clusters')

plt.show()
```

How many clusters are there?

```python
km3 = KMeans(3)
km3.fit(data_scaled)
data3 = pd.DataFrame(data_scaled)
data3['cluster'] = km3.fit_predict(data_scaled)
```

```python
plt.scatter(data3[0], data3[1], c=data3['cluster'], cmap='rainbow')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
km5 = KMeans(5)
km5.fit(data_scaled)
data5 = pd.DataFrame(data_scaled)
data5['cluster'] = km5.fit_predict(data_scaled)
```

```python
plt.scatter(data5[0], data5[1], c=data5['cluster'], cmap='rainbow')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```
