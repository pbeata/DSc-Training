```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('../data/countries_elbow_exercise.csv')
data
```

```python
plt.scatter(data['Longitude'], data['Latitude'])
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

```python
# choose the data for clustering
x = data[['Longitude', 'Latitude']]
```

```python
# perform clustering for 4 clusters initially
kmeans = KMeans(4)
kmeans.fit(x)
clusters = kmeans.fit_predict(x)
data['Cluster'] = clusters
```

```python
# plot the results by cluster
plt.scatter(data['Longitude'], data['Latitude'], c=data['Cluster'], cmap='rainbow')
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

### Selecting Optimal Number of Clusters

```python
wcss = kmeans.inertia_
wcss
```

```python
# calculate the WCSS for any number of clusters up to N

N = 10
WCSS = []
for i in range (1, N+1):
    kmeans = KMeans(i)
    kmeans.fit(x)
    WCSS.append(kmeans.inertia_)
```

```python
WCSS
```

```python
n = np.arange(1, N+1)
n
```

```python
plt.figure(figsize=(6,3), dpi=100)

plt.plot(n, WCSS, 'ko-')
plt.ylabel('WCSS')
plt.xlabel('Clusters')

plt.show()
```

```python
n = len(WCSS)

w_old = WCSS[0]
for i in range (1, n):
    w = WCSS[i]
    diff = w - w_old
    diff_percent = 100 * diff / w_old
    print(diff_percent)
    w_old = w
```
