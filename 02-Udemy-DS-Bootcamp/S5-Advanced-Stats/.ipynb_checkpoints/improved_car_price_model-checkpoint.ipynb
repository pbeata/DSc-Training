### Dummy Variables

```python
data_with_dummies = pd.get_dummies(data_no_mc.drop(['Model'], axis=1), drop_first=True)
# data_with_dummies = pd.get_dummies(data_no_mc, drop_first=True)

data_with_dummies
```

```python
# variance inflation factors
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_vars = data_with_dummies.drop(['Price'], axis=1)
# vif_vars = vif_vars.drop(['Model'], axis=1)
vif_all = pd.DataFrame()

vif_all['VIF'] = [variance_inflation_factor(vif_vars.values, i) for i in range(vif_vars.shape[1])]
vif_all['Features'] = vif_vars.columns

vif_all.sort_values('VIF')
# vif_all.sort_values('Features')
# vif_all
```

```python
vif_all[vif_all['VIF'] < 5].sort_values('VIF')
```

```python
data = data_with_dummies
# data = data.drop(['EngineV'], axis=1)
```

### Linear Regression Model

```python
# declare our inputs and targets
# target = data['Price']
target = np.log(data['Price'])
inputs = data.drop(['Price'], axis=1)
```

```python
# scale the inputs 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(inputs)
xs = scaler.transform(inputs)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(xs, target, test_size=0.2, random_state=101)
```

### Create the Regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.figure()
plt.scatter(y_train, y_hat)
plt.show()
```

```python
sns.displot(y_train - y_hat)
```

```python
reg.score(x_train, y_train)
```

```python
def compute_adjusted_R2(x, R2):
    n = x.shape[0]
    p = x.shape[1]
    c = (n - 1) / (n - p - 1)
    result = 1 - (1 - R2) * c
    return result.round(3)
```

### Weights and Bias

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

### Testing

```python
yhat_test = reg.predict(x_test)
```

```python
plt.figure()
plt.scatter(y_test, yhat_test, alpha=0.2)
plt.show()
```

```python
df_perf = pd.DataFrame(np.exp(yhat_test), columns=['Prediction'])
# df_perf = pd.DataFrame(yhat_test, columns=['Prediction'])
df_perf
```

```python
y_test = y_test.reset_index(drop=True)
df_perf['Target'] = np.exp(y_test)
# df_perf['Target'] = y_test
df_perf
```

```python
df_perf['Residual'] = df_perf['Target'] - df_perf['Prediction']
df_perf
```

```python
df_perf['Diff. Percent'] = 100 * np.absolute(df_perf['Residual'] / df_perf['Target'])
df_perf.describe()
```

```python
pd.options.display.max_rows = 999

df_perf.sort_values(by=['Diff. Percent'], ascending=False)
```
