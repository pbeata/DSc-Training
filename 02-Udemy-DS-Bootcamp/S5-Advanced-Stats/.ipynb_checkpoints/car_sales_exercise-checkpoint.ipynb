```python
# PART 1

data_preprocessed.columns
```

```python
vif_vars = data_preprocessed[['log_price', 'Mileage', 'EngineV', 'Brand_BMW', 'Brand_Mercedes-Benz',
       'Brand_Mitsubishi', 'Brand_Renault', 'Brand_Toyota', 'Brand_Volkswagen',
       'Body_hatch', 'Body_other', 'Body_sedan', 'Body_vagon', 'Body_van',
       'Engine Type_Gas', 'Engine Type_Other', 'Engine Type_Petrol',
       'Registration_yes']]
```

```python
vif_all = pd.DataFrame()
```

```python
vif_all['VIF'] = [variance_inflation_factor(vif_vars.values, i) for i in range(vif_vars.shape[1])]
```

```python
vif_all['Features'] = vif_vars.columns
```

```python
vif_all.sort_values('VIF')
```

By calculating the variance inflation factors, we can see that the majority of the features are in the good range of 1 to 5. However, there are two features (Engine Volume and Registration Yes) that are 13.4 and 15.2, respectively. 

Additionally, the one strange outlier is the log_price because its VIF is 42. However, the log_price is NOT a feature: it is the dependent variable we are trying to predict/model.

ANSWER: 

Obviously, 'log_price' has a very high VIF. This implies it is most definitely **linearly correlated** with all the other variables. And this is no surprise! We are using a linear regression to determine 'log_price' given values of the independent variables! This is exactly what we expect - a linear relationship!

However, to actually assess multicollinearity for the predictors, we have to drop 'log_price'. The multicollinearity assumption refers only to the idea that the **independent variables** shoud not be collinear.

```python
# Part 2
# Compute the VIF but this time without the dependent variable "log_price"

vif_vars = data_preprocessed.drop(['log_price'], axis = 1)
vif_vars
```

```python
vif_all = pd.DataFrame()
vif_all['VIF'] = [variance_inflation_factor(vif_vars.values, i) for i in range(vif_vars.shape[1])]
vif_all['Features'] = vif_vars.columns

# vif_all.sort_values('VIF')
vif_all.sort_values('Features')
```

The features EngineV and Registration Yes have higher values than the acceptable range of 1 < VIF < 5, so we can likely drop these two features from the model. 

```python
# Part 3
# Now, use the get_dummies method but do not drop the first feature.

# data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
data_with_dummies = pd.get_dummies(data_no_multicollinearity)
data_with_dummies.head()
```

```python
vif_vars = data_with_dummies.drop(['log_price'], axis = 1)
vif_all = pd.DataFrame()

vif_all['VIF'] = [variance_inflation_factor(vif_vars.values, i) for i in range(vif_vars.shape[1])]

vif_all['Features'] = vif_vars.columns
vif_all

# vif_all.sort_values('VIF')

# vif_vars
```
