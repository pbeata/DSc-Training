### Shoe Sales Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

We have sales data by shoe size from two years: 2015 and 2016. 

```python
usa15 = pd.read_csv('./data/shoes_2015.csv')
usa15
```

```python
usa16 = pd.read_csv('./data/shoes_2016.csv')
usa16
```

```python
usa15
```

```python
data = usa15.copy()

col_list = data.columns
month = int(13)
for col in col_list:
    if (col != 'US'):
        data[str(month)] = usa16[col]
        month += 1

data
```

```python
# compute the mean sales for each shoe size

data['Mean_1516'] = data.iloc[:, 1:25].mean(axis=1)
data
```

```python
from math import sqrt

# determine the sample size
n = 24

# determine the standard deviation by row
data['StDev'] = data.iloc[:, 1:25].std(axis=1)

# compute the standard error:  s / sqrt(n)
data['SE'] = data['StDev'] / sqrt(n)

# determine the t-statistic 
tstat = 2.074

# compute the margin of error
data['ME'] = tstat * data['SE']

# compute the 95% CI for the 2015-16 data
data['CI_start'] = data['Mean_1516'] - data['ME']
data['CI_end'] = data['Mean_1516'] + data['ME']


data
```

```python
# round the upper limit of the CI to determine how many pairs of shoes we should order
data['Number_Pairs'] = data['CI_end'].round(0).astype(int)
data
```
