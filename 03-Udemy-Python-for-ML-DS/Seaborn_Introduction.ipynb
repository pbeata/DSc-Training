### Comparison Plots

* jointplot(): histogram with scatterplot, 2D KDE shaded distribution
* pairplot(): common in EDA of ML

#### Pair Plot

* It gives a quick way to compare all numerical columns in a DataFrame
* It automatically creates a histogram for each column and a scatterplot comparison between all possible combinations of columns
* WARNING: this can be a bit CPU and RAM intensive

```python
df.head()
```

```python
sns.jointplot(data=df, x='math score', y='reading score');
```

```python
sns.jointplot(data=df, x='math score', y='reading score',
             kind='hex');
```

```python
sns.jointplot(data=df, x='math score', y='reading score',
             alpha=0.2);
```

```python
sns.jointplot(data=df, x='math score', y='reading score',
             kind='hist');
```

```python
sns.jointplot(data=df, x='math score', y='reading score',
             kind='kde', shade=True);
```

```python
sns.jointplot(data=df, x='math score', y='reading score',
             hue='gender');
```

```python
sns.pairplot(data=df, hue='gender');
```

```python
sns.pairplot(data=df, hue='gender', diag_kind='hist');
```

```python
sns.pairplot(data=df, hue='gender', corner=True);
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box',
           row='lunch');
# similar to using hue to split by category
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box',
           col='lunch');
# similar to using hue to split by category
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box',
           col='lunch', row='test preparation course');
# try to limit this to 2x2 grid only
```

```python
sns.pairplot(df);
```

```python
# manually created your pair grid

g = sns.PairGrid(df, hue='gender')

g = g.map_upper(sns.scatterplot)

# g = g.map_diag(sns.histplot)
g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend()
```

### Matrix Plots

* visual equivalent of displaying a pivot table
* two main types: heatmap and clustermap

The heatmap() should ideally have all cells be in the same UNITS, so the color mapping will make sense across the entire DataFrame.

```python
df = pd.read_csv('./data/country_table.csv')
df
```

```python
df = df.set_index("Countries")
df  # now this is a good candidate for heatmap()
```

```python
sns.heatmap(df);
# here, we include YEARS and RATE units (wrong!)
```

```python
# plt.figure(figsize=(12,8), dpi=300)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5,
           annot=True);
```

```python
# color mapping for colorblind people is important
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5,
           annot=True,
           cmap='viridis');

# change the colorbar with the center=x keyword argument
```

```python
sns.clustermap(df.drop('Life expectancy', axis=1), linewidth=0.5,
           annot=True,
           cmap='viridis');
```

```python
sns.clustermap(df.drop('Life expectancy', axis=1), linewidth=0.5,
           annot=True,
           cmap='viridis',
           col_cluster=False);
```
