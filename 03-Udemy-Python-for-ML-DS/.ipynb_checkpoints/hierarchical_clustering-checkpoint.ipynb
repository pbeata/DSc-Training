## Hierarchical Clustering - Part 1

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%config Completer.use_jedi = False
```

```python
df = pd.read_csv('./data/cluster_mpg.csv')
df
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_with_dummies = pd.get_dummies(df.drop('name', axis=1))
df_with_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# use the MinMaxScaler to scale each to feature to the range [0, 1]
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_with_dummies)
```

```python
scaled_data.min()
```

```python
scaled_data.max()
```

```python
df_scaled = pd.DataFrame(data=scaled_data, columns=df_with_dummies.columns)
df_scaled
```

```python
plt.figure(figsize=(10,6), dpi=100)
sns.heatmap(df_scaled, cmap='rainbow');
```

```python
sns.clustermap(df_scaled, row_cluster=False);
```

```python
df_scaled.corr()
```

```python
sns.heatmap(df_scaled.corr(), cmap='viridis');
```
