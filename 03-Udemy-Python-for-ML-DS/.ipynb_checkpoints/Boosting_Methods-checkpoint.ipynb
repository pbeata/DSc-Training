## The Data

### Mushroom Hunting: Edible or Poisonous?

Data Source: https://archive.ics.uci.edu/ml/datasets/Mushroom


This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like ``leaflets three, let it be'' for Poisonous Oak and Ivy.


Attribute Information:

4. bruises?: bruises=t,no=f

6. gill-attachment: attached=a,descending=d,free=f,notched=n
7. gill-spacing: close=c,crowded=w,distant=d
8. gill-size: broad=b,narrow=n
9. gill-color: black=k,brown=n,buff=b,chocolate=h,gray=g, green=r,orange=o,pink=p,purple=u,red=e, white=w,yellow=y
10. stalk-shape: enlarging=e,tapering=t
11. stalk-root: bulbous=b,club=c,cup=u,equal=e, rhizomorphs=z,rooted=r,missing=?

20. spore-print-color: black=k,brown=n,buff=b,chocolate=h,green=r, orange=o,purple=u,white=w,yellow=y

```python
X = df.drop('class', axis=1)
```

```python
X.isnull().sum()
```

```python
# convert categorical data to dummy variables
X = pd.get_dummies(X, drop_first=True)
X
```

```python
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

## The Model

```python
# the AdaBoostClassifier is typical used with a decision tree classifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=1)  # =1 means a SINGLE stump
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score
```

```python
y_pred =  model.predict(X_test)
```

```python
y_pred
```

```python
print(classification_report(y_test, y_pred))
```

```python
model.feature_importances_

# here there is a single feature being split on:
```

```python
model.feature_importances_.argmax()
```

```python
X.columns[22]  # so the single feature we would look for in the field would be ODOR (odor n)
```

```python
sns.countplot(data=df, x='odor', hue='class');
# most likely that if the mushroom does not have an odor, it is okay to eat it
```

```python
len(X.columns)
```

```python
error_rates = []

for n in range(1, 96):
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    err = 1 - accuracy_score(y_test, y_pred)
    error_rates.append(err)
```

```python
plt.plot(range(1,96), error_rates, 'ko-');
```

```python
model.feature_importances_
```

```python
features = pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=['Importance'])
features
```

```python
important_features = features[ features['Importance'] > 0 ]
important_features
```

```python
plt.figure(figsize=(14,6), dpi=200)
sns.barplot(data=important_features.sort_values("Importance"), 
            x=important_features.index,
            y="Importance")
plt.xticks(rotation=90)
plt.title("Importance of Each Factor using 95 Estimators")

plt.show()
```

```python
model = AdaBoostClassifier(n_estimators=19)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
err = 1 - accuracy_score(y_test, y_pred)
print(err)
```

```python
features = pd.DataFrame(index=X.columns, 
                        data=model.feature_importances_, 
                        columns=['Importance'])
important_features = features[ features['Importance'] > 0 ]
```

```python
plt.figure(figsize=(14,6), dpi=200)
sns.barplot(data=important_features.sort_values("Importance"), 
            x=important_features.index,
            y="Importance")
plt.xticks(rotation=90)
plt.title("Importance of Each Factor using 19 Estimators")

plt.show()

# remember, we can use grid search to find best number of estimators
```

## Gradient Boosting

```python
df = pd.read_csv('./data/mushrooms.csv')
```

```python
X = df.drop('class', axis=1)
X = pd.get_dummies(X, drop_first=True)
```

```python
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# sets up for decision tree by default with max_depth default of 3 --> should use Grid Search
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators': [50, 100],
              'learning_rate': [0.1, 0.05, 0.2],
              'max_depth': [3, 4, 5]}
```

```python
gb_model = GradientBoostingClassifier()
```

```python
grid = GridSearchCV(gb_model, param_grid)
```

```python
grid.fit(X_train, y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
grid.best_params_
```

```python
print(classification_report(y_test, y_pred))
```

```python
# feature importance taken from the actual best model
grid.best_estimator_.feature_importances_
```

```python
feat_imp = grid.best_estimator_.feature_importances_ 
```

```python
imp_feat = pd.DataFrame(index=X.columns,
                        data=feat_imp,
                        columns=["Importance"])
```

```python
imp_feat
```

```python
imp_feat = imp_feat[ imp_feat['Importance'] > 0.0005 ]
```

```python
imp_feat
```

```python
imp_feat = imp_feat.sort_values('Importance')
```

```python
plt.figure(figsize=(12,8), dpi=200)

sns.barplot(data=imp_feat,
            x=imp_feat.index,
            y='Importance')
plt.xticks(rotation=90)

plt.show()
```
