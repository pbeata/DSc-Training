### Matplotlib Subplots

```python
# FORMAT: plt.subplots(num_rows=m, num_cols=n)

a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes.plot(x, y);  # if there is only (1x1) plots

axes[0].plot(x, y);
axes[1].plot(a, b);
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

# axes.plot(x, y);  # if there is only (1x1) plots

axes[0][0].plot(x, y);
axes[1][0].plot(a, b);

axes[0][1].plot(x, y);
axes[1][1].plot(a, b);
```

```python
axes
```

```python
axes.shape
```

```python
axes[0][0]
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(x,y);

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)

# plt.tight_layout()

# xspace is based on the average axes size
fig.subplots_adjust(wspace=0.5, hspace=0.5)

axes[1][0].set_ylabel('Sample Y-Label')

axes[1][1].set_title('This is the Title')

# figure level title
fig.suptitle("Figure-Level Title", fontsize=12)

plt.show()

fig.savefig('./out/subplots_example.png', bbox_inches='tight', dpi=300)
```

### Styling in Matplotlib

```python
x = np.linspace(0, 11, 10)
y = x ** 2
z = np.sqrt(x)
```

```python
# Part 1: Legend

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label='$y(x) = x$')
ax.plot(x, y, label='$y(x) = x^2$')

# ax.legend();
# ax.legend(loc=9);
ax.legend(loc=(1.1, 0.5));  # move legend to the outside of the plot
```

```python
# Part 2: Colors and Line Styles

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# lw = linewidth
# ls = linestyle

ax.plot(x, x, color='blue', lw=4, ls=':');     # basic color string input
ax.plot(x, y, color='#ababab', lw=0.4, ls='--', marker='^');  # RGB HEX code as color input
ax.plot(x, z, color='red', lw=1, ls='-.', marker='+', markersize=10);
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='magenta', lw=2, ls='--',
        marker='o', 
        markersize=10, 
        markerfacecolor='red', 
        markeredgewidth=2, 
        markeredgecolor='black');
```
