```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%config Completer.use_jedi = False
```

```python
df = pd.read_csv('./data/cluster_mpg.csv')
df
```

```python
df_with_dummies = pd.get_dummies(df.drop('name', axis=1))
df_with_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# use the MinMaxScaler to scale each to feature to the range [0, 1]
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_with_dummies)
```

```python
df_scaled = pd.DataFrame(data=scaled_data, columns=df_with_dummies.columns)
df_scaled
```

## Hierarchical Clustering - Part 2

Clusters and Dendrograms

```python
sns.clustermap(df_scaled, row_cluster=True);
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# if you already have an intuition of the number of clusters in your data
model = AgglomerativeClustering(n_clusters=3)
```

```python
cluster_labels = model.fit_predict(df_scaled)
cluster_labels
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_labels, palette='viridis')

plt.show()
```

```python
# affinity: default is fine, Euclidean distance; it has a nice intuitive understanding when using MinMaxScaler

# what is the max distance possible?
#  sqrt(N_features)

N_features = len(df_scaled.columns)
np.sqrt(N_features)  # MAX POSSIBLE DISTANCE BETWEEN TWO POINTS
```

```python
model = AgglomerativeClustering(
    n_clusters=None, 
    distance_threshold=0  # this will assign a unique cluster label to every data point 
)
```

```python
cluster_labels = model.fit_predict(df_scaled)
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
linkage_matrix
```

```python
plt.figure(figsize=(20,10), dpi=100)
dendro = hierarchy.dendrogram(linkage_matrix)
```

```python
plt.figure(figsize=(20,10), dpi=100)
dendro = hierarchy.dendrogram(linkage_matrix,
                              truncate_mode='lastp',
                              p=10)
```

```python
plt.figure(figsize=(20,10), dpi=100)
dendro = hierarchy.dendrogram(linkage_matrix,
                              truncate_mode='level',
                              p=3)
```

```python
# another example with higher distance threshold
model = AgglomerativeClustering(
    n_clusters=None, 
    distance_threshold=2  # this will assign a unique cluster label to every data point 
)

# compute cluster labels
cluster_labels = model.fit_predict(df_scaled)

# update the linkage matrix
linkage_matrix = hierarchy.linkage(model.children_)

# plot the new dendrogram
plt.figure(figsize=(20,10), dpi=100)
dendro = hierarchy.dendrogram(linkage_matrix,
                              truncate_mode='lastp',
                              p=11)
```
